```python
# Can you plot a bar chart of distribution where it shows the year when movies are released? What year had the most released movies?

# Find the number times a movie was released during a specific year for each year.
df_year = df['Year'].value_counts()

# Just determining the size of the image
fig, ax = plt.subplots(figsize=(20,10))

# Now I'm plotting the bar using index and values of df_year
ax.bar(df_year.index, df_year.values)

# I kept having an overlapped values in the x axis so I adjusted them so it is easy to read.
ax.set_xticks([1900, 1950, 2000, 2020])
ax.set_xticklabels([1900, 1950, 2000, 2020])

# Just labeling the x and y axis and adding a title.
ax.set_xlabel('Year')
ax.set_ylabel('Number of Movies')
ax.set_title('Number of Movies Released by Year')

# This command will show the bar chart
plt.show()



print('It appears that around 2017 was the year that had the most released movies.')
```

```python
# Save as an image (png only)! Call it MoviesByYear

plt.savefig('MoviesByYear.png')
```

```python
# How many movies were released during that specific popular year?

specific_year = df[df['Year'] == 2017]

specific_year.count()

# It looks like 1401 movies were released in 2017 (The highest compared to the rest).
```

```python
# Can you plot a horizontal bar chart of distribution of different content ratings (Age).

# Nothing big, we can just use value count and graph that variable.
df_ages = df['Age'].value_counts()

fig, ax = plt.subplots(figsize=(20,10))


df_ages.plot(kind='barh')


plt.xlabel('Number of Movies')
ax.set_ylabel('Content Ratings')
ax.set_title('Number of Movies Released by Content Ratings')


plt.show()
```

```python
# Please create a PivotTable presenting the average IMDb scores of titles on Netflix, arranged by content rating group.

df_IMDb = df.pivot_table(values="IMDb", index="Age", columns = "Netflix", aggfunc='mean')
df_IMDb.head()
```

```python
# Can you line chart that shows runtime by year starting from 1950 and ending in 2005?


year1 = df[df['Year'].isin(list(range(1950,2006)))]
runntime = year1['Runtime']

plt.plot(year1['Year'].astype(int), runntime, marker='o')

plt.xlabel('Year')
plt.ylabel('Runtime (mins)')
plt.title('Runtime by Year')

plt.show()
```

```python
# What is the maximum runtime of all time? What year was it?

year_max_runtime = year1[year1['Runtime'] == year1['Runtime'].max()]['Year']
max_runtime = year1['Runtime'].max()


print(year_max_runtime) #1964
print(max_runtime) #750
```

```python
# Drop the "Type" column

df = df.drop('Type', axis=1)
df.head()
```

```python
# Sum up the scores for Netflix, Hulu, Prime Video and disney+ (==1) for each content rating group (Age).

# There are two values in these columns, 1 and 0 which means its available on that streaming service and its not available on that streaming service.
df_1 = df[(df['Netflix'] == 1) | (df['Hulu'] == 1) | (df['Prime Video'] == 1) | (df['Disney+'] == 1)]
df_All = df_1.groupby("Age").agg({"Netflix": "sum", "Hulu": "sum", "Prime Video": "sum", "Disney+": "sum"})
df_All[[ 'Netflix', 'Hulu', 'Prime Video', 'Disney+']].head()
```
